```python
## Q44. Create a pivot table displaying average LOS based on Service and Primary Diagnosis.

#Answer

#Linking with the dataframe
ReAdmissionRegistry = pd.read_excel("HospitalDatabase.xlsx","ReAdmissionRegistry")

#Creating the pivot table
pivot = np.round(pd.pivot_table(ReAdmissionRegistry, values='ExpectedLOS', 
                                index='Service', 
                                columns='PrimaryDiagnosis', 
                                aggfunc=np.mean),2)

#Displaying the pivot table
display(pivot)

# The result shows the pivot table displaying the Length of Stay of patients based on the type of service and primary diagnosis
```

```python
## Q48. Display the average LOS based on Service and Discharge Disposition(Maintain the column Hierarchy).

#Linking to the dataframe
Discharges = pd.read_excel("HospitalDatabase.xlsx","Discharges")

# Setting the index to maintain column hierarchy according to the dataset
df = Discharges
df.set_index(['DischargeDisposition','Service', 'ExpectedLOS'])

#Using the pivot table to calculate average Length of Stay 
#based on Service and Discharge Disposition
pivot = np.round(pd.pivot_table(Discharges, values='ExpectedLOS', 
                                index='Service', 
                                columns='DischargeDisposition', 
                                aggfunc=np.mean),2)

#Displaying the pivot table
display(pivot)
```

```python
## Q50. Which departmentID has maximum number of visits scheduled.

#Importing Counter and linkinf the dataframe
from collections import Counter
df = pd.read_excel('HospitalDatabase.xlsx', sheet_name = 'AmbulatoryVisits')

#Finding the DepartmentID with maximum number of visits scheduled
entry = df['VisitDepartmentID'].tolist()
id_counts = Counter(entry)
most_common_id = id_counts.most_common(1)[0][0]

# Displaying the result
print(f'The Department ID with maximum number of scheduled visits is {most_common_id}.')
```
